Q.4 In which `Company Location` is 'Remote' `Work Setting` the most common? What is the most common `Employee Residence` for such employees?

```python
df.head()
```

```python
plt.pie(df['work_setting'].value_counts(), labels = df['work_setting'].value_counts().index)
```

```python
remote_employees = df[df['work_setting'] == 'Remote']
location_counts = remote_employees['company_location'].value_counts()
most_common_location = location_counts.idxmax()
most_common_location_employees = remote_employees[remote_employees['company_location'] == most_common_location]
residence_counts = most_common_location_employees['employee_residence'].value_counts()
most_common_residence = residence_counts.idxmax()
print(f"In the company location '{most_common_location}', 'Remote' work setting is the most common.")
print(f"The most common employee residence for such employees is '{most_common_residence}'.")
```

Q.6 What is the ratio of `Employment Type`?

```python
df.head()
```

```python
plt.figure(figsize=(8,8))
plt.pie(df['employment_type'].value_counts(), labels = df['employment_type'].value_counts().index,autopct='%1.1f%%')
plt.legend(df['employment_type'].value_counts().index, title='Employment Type', loc='upper right')
plt.show()
```

`Full-time` = 99.5%
`Contract` = 0.2%
`Part-time` = 0.2%
`Freelance` = 0.1%

From this dataset we can conclude that:
1. Machine Learning and AI is the highest paying Job Category in the Data Sector.
2. The most common Employment Type is Full Time.There is also a huge gap between the number of Full Time employees and, other Part Time,Contractual or Freelancing employees. This shows that there is a perennial demand for Data Related Services which results in Full Time employment opportunities.
3. The most number of Data Professionals work in company based in The US. Moreover, US based companies also employ the most number of Remote Workers. This shows the popularity of Data Related Services in The US Market.
4. We also observe that there is a great disparity between the United States and other countries based on the employment status of Data Professionals. This might be due to the absence/lack of Data Companies in other countries or migration of Data Professionals to the US due to higher pay in US.
5. Data Science and Research employs the most number of Remotely Working Data Professionals.
